### Generators
- a fun which yields many fields at a time

```python
nums=list(map(int,input().split()))
nums
```

```python
def length(it):
    return len(it)
lns=tuple(map(length,['apple','banana','orange','brinjal']))
lns
```

```python
#applying lambda on iterable using map
evens=list(map(lambda x:x*2,[4,5,61,7,8,1]))
evens
```

```python
def product(x,y):
    return x*y

li=[3,4,67,8]
reduce(product,li)
```
