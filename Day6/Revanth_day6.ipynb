```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
iris=load_iris()
iris
```

```python
df=pd.DataFrame(iris['data'],columns=iris.feature_names)
df['Target']=iris.target
#df['flower']=iris[iris.target_names[df['Target']]]
df.sample(10)
```

```python
x=df.drop('Target',axis=1)
y=df['Target']
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x,y)
```

```python
knn.predict([[6.2,2.2,4.5,1.5]])
```

```python
from sklearn.metrics import accuracy_score
y_p=knn.predict(x)
accuracy_score(y,y_p)
```

```python
class car:
    def __init__(self,name):
        self.name = name
    def display(self):
        print('My car is :',self.name)
```

```python
class student:
    def __init__(self,name,rolno,branch):
        self.name=name
        self.rolno=rolno
        self.branch=branch
    def details(self):
        print('Student name is',self.name)
        print('student rollno is',self.rolno)
        print('student branch is',self.branch)
```

```python
s1=student('revanth','y20cb037','CSBS')
```

```python
s1.details()
```

```python
#inheritence
class parent:
    def __init__(self):
        self.name='rahul'
    def display(self):
        print('I am from class parent')
class child(parent):
    def __init__(self):
        self.age=19
    def dis(self):
        print('I am from class child')
        print('I am',self.age,'year old')
        
```

```python
cl=child()
cl.display()
cl.dis()
```

```python
#multiple inheritence
class main:
    def __init__(self):
        self.st='python'
    def display(self):
        print('I am from class main')

class second:
    def __init__(self):
        self.st2='Data Analtsis'
    def display2(self):
        print('I am from class second')
class third:
    def __init__(self):
        self.st='Machine Learning'
        main.__init__(self)
        second.__init__(self)
    def display3(self):
        print(self.st3)
        print9"i am from class third"
```
