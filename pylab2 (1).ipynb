```python
#programme o find the number of days in a month
print('Enter the month number')
big=[1,3,5,7,8,10,12]
small=[4,6,9,11]
month=int(input())
if month in big:
  print(31,'days')
elif month in small:
  print(30,'days')
else: print(28,'days')
```

```python
#program to encrypt/decrypt the password
def encrypt(s):
    encrypt_pwd=''
    for ch in s:
        encrypt_pwd+=chr(ord(ch)+5)
    return encrypt_pwd
def decrypt(s):
    decrypt_pwd=''
    for ch in s:
        decrypt_pwd+=chr(ord(ch)-5)
    return decrypt_pwd

pwd=input('Enter yor password: ')
print('encrypted password is',encrypt(pwd))
print('decrypted password is',decrypt(encrypt(pwd)))
```

```python
#programme to fint frequency of the word
s=input('Enter the string: ')
w=input('Enter the word to search: ')
count=0
wl=len(w)
for i in range(len(s)-len(w)+1):
    if w == s[i:wl]:
        count+=1
    wl+=1
print(w,'repeated',count,'times in',s)
```

```python
#temparature converter
def fahr(t):
    c=round(((t-32)*5)/9,2)
    return c
def celc(t):
    f=round(((9*t)/5)+32,2)
    return f
temp=float(input())
print('temparature in celcius:',celc(temp),'C')
print('temparature in celcius:',fahr(temp),'F')
```
